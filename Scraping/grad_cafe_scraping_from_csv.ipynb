```python
import pandas as pd          
```

```python
# read specific columns of csv file using Pandas 
col_Names=['University','Accepted']
df= pd.read_csv("all_i_400-600.csv",names=col_Names)
print(df)
print(df.size)
```

```python
list_of_acceptance=df['Accepted'].tolist()
for i in range(len(list_of_acceptance)):
    list_of_acceptance[i]=list_of_acceptance[i].replace("\\n\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t","")
    list_of_acceptance[i]=list_of_acceptance[i].split(" ")[0][2:]
# print(list_of_acceptance)
df['result']=list_of_acceptance
print(df)
```

```python
list_of_scores=df['Accepted'].tolist()
# print(list_of_scores)
for i in range(len(list_of_scores)):
    list_of_scores[i]=list_of_scores[i].replace("\\n\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t","")
    list_of_scores[i]=list_of_scores[i].split(" ")[1:]
        
df['scores']=list_of_scores
print(df)
```

```python
final_scores=[]
for details in list_of_scores:
    scores=[]
    if(len(details)!=0):
        if(details[0]=='on'):
                scores+=details[3:]
                final_scores.append(scores)
        else:
            scores+=details
            final_scores.append(scores)
    else:
        scores+=details
        final_scores.append(scores)
```

```python
# print(scores)
df['details']=final_scores   
df=df.drop(['scores'], axis=1)
df=df.drop(['Accepted'], axis=1)
df = df[df.result != 'Rejected']
df = df[df.result != 'Interview\',']
df = df[df.result != 'Wait']
df = df[df.result != 'Other\',']

print(df)
```

```python
scores_list=df['details'].tolist()
scores_list[:5]
```

```python
term=[]
details=df['details'].tolist()
for i in range(len(scores_list)):
    # print(len(scores_list[i]))
    if len(scores_list[i])>=2:
#         print(scores_list[i][0]+" "+scores_list[i][1])
        if(scores_list[i][0]=='\'Fall' or scores_list[i][0]=='\'Spring'):
            term.append(scores_list[i][0]+" "+scores_list[i][1])
#             print(scores_list[i][0]+" "+scores_list[i][1])
        elif(scores_list[i][1]=='\'Fall' or scores_list[i][1]=='\'Spring'):
            term.append(scores_list[i][1]+" "+scores_list[i][2])
#             print(scores_list[i][1]+" "+scores_list[i][2])
        else:
            term.append(None)
            
    else:
        term.append(None)
df['term']=term
df
```

```python
columns_required=set()
for i in range(len(details)):
    for j in range(len(details[i])):
        # print(details[i][j])
        details[i][j] = details[i][j].replace("'","")
        details[i][j] = details[i][j].replace(",","")
        # print(details[i][j])
        if details[i][j].isalpha():
            columns_required.add(details[i][j])

columns_required
```
