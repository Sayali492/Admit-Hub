```python
import matplotlib.pyplot as plt
import requests
import requests
from urllib.request import urlopen
import re
# from time import sleep
# from IPython.core.debugger import Tracer
# Bowlie- ,Northeastern University-Seattle-143, Columbus State University-19 ,Clark University-25

def find_n_pages(html):
    reg = re.search('Results\s\d+\s-\s\d+ of\s\d+',html)
    string=reg.group()
#     print(string)
    regex = r"(\d+)"
    match = re.findall(regex, string)  
    return match[2]

for unis in range(474,601):
    print("Getting data of "+ universities[unis] )
    DATA_DIR = './university/'
    url_form = "https://www.thegradcafe.com/survey/?per_page=40&q="+universities[unis]+"&institution=&program=&degree=&page=0"
    url_form = url_form.replace(" ", "%20")
    response = urlopen(url_form)
    html = response.read().decode('utf-8')
    n=find_n_pages(html)
#     print(n)
    if(int(n)>15000):
        n=0
    no_of_pages=int(n)/40
    pages=int(no_of_pages) +1
    university_pages[universities[unis]]=pages
#     print(pages)
    for i in range(1,pages+1):
        url_form = "https://www.thegradcafe.com/survey/?per_page=40&q="+universities[unis]+"+&institution=&program=&degree=&page={}"

        url = url_form.format(i)
        r = requests.get(url)
        fname = "{data_dir}/{page}.html".format(
          data_dir=DATA_DIR,
          page=universities[unis]+str(i)
        )
        with open(fname, 'wb') as f:
            f.write(r.text.encode('UTF-8'))
    #         sleep(1)
            print("getting {0}...".format(i))
```

```python
from bs4 import BeautifulSoup
import datetime
import time
# from IPython.core.debugger import Tracer
import sys
import re
import pandas as pd
import numpy as np
import os

data = []
unis_list=[]
combined=[]
df = pd.DataFrame()
directory="./university"
for unis in range(400,601):
     print(unis)
     for i in range(1,3000):
        path=directory+"/"+universities[unis]+str(i)+".html"
        try:
            with open(path, 'rb') as f:
                    soup = BeautifulSoup(f.read())
                    tables = soup.findAll('div', class_='row mb-2')
            #         count = 1
                    for tab in tables:
                        details=tab.findAll('div',class_='col')
                        for data in details:
                            unis_=data.find('h6',class_='mt-3 fw-normal')
                            others=data.findAll('span',class_='badge')
                            if unis_!=None:
                                unis_list.append(unis_.text)
                            else:
                                unis_list.append("NULL")
                            data=[]
                            for d in others:
                                data.append(d.text.strip())
                            combined.append(data)
        except:
            continue
            
df['university']=unis_list
df['data']=combined

df.to_csv('all_i_400-600.csv', header=None)
```
