```python
import numpy as np
from scipy.stats import f_oneway

def calculate_anova_fvalue(feature_values, target_variable):
    groups = {}
    for value, target in zip(feature_values, target_variable):
        if target not in groups:
            groups[target] = []
        groups[target].append(value)
    
    # Calculate ANOVA F-value
    f_value, _ = f_oneway(*groups.values())
    return f_value

# Sample dataset
# Each row represents a data point, and each column represents a feature
# The last column is the target variable

# Extract features and target variable
correlation_matrix = df[['toeflScore', 'GRE_AWA', 'GRE_QUANT', 'GRE_VERBAL', 'GPA',
       'Publications', 'Research Experience', 'Work experience', 'Degree Encoded', 'Course Name Encoded']]

# Extract the correlation of 'university_name_encoded' with all other features

features = correlation_matrix
target_variable = df['University Name Encoded']

# Calculate ANOVA F-value for each feature
anova_f_values = []
for i in range(features.shape[1]):
    feature_values = features.iloc[:, i]
    f_value = calculate_anova_fvalue(feature_values, target_variable)
    anova_f_values.append(f_value)

# Assign feature weights based on ANOVA F-values
feature_weights = anova_f_values

print("Feature Weights:",correlation_matrix.columns, feature_weights)
```

```python
import numpy as np

def calculate_gini_index(feature_values, class_labels):
    # Combine feature values and corresponding class labels
    data = list(zip(feature_values, class_labels))
    total_samples = 10
    # Sort the data based on feature values
    data.sort()

    min_gini_index = float('inf')
    split_point = None

    # Iterate through possible split points
    for i in range(1, total_samples):
        left_data = data[:i]
        right_data = data[i:]

        # Calculate Gini index for left and right partitions
        left_count = len(left_data)
        right_count = len(right_data)

        left_gini = gini_index([label for _, label in left_data])
        right_gini = gini_index([label for _, label in right_data])

        # Calculate weighted average Gini index
        weighted_gini = (left_count / total_samples) * left_gini + (right_count / total_samples) * right_gini
        print(i,weighted_gini)

        # Update minimum Gini index and split point if necessary
        if weighted_gini < min_gini_index:
            min_gini_index = weighted_gini
            split_point = i

    return min_gini_index, split_point

def gini_index(class_labels):
    counts = {}
    total_samples = len(class_labels)

    # Count the frequency of each class label
    for label in class_labels:
        if label in counts:
            counts[label] += 1
        else:
            counts[label] = 1

    # Calculate the Gini index
    gini = 1.0
    for label in counts:
        probability = counts[label] / total_samples
        gini -= probability ** 2

    return gini

def normalize_gini_index(gini_values):
    min_gini = min(gini_values)
    max_gini = max(gini_values)
    normalized_gini = [(gini - min_gini) / (max_gini - min_gini) for gini in gini_values]
    return normalized_gini


# Extract features and class labels
# features = df.iloc[:, :-1]
# class_labels = df.iloc[:, -1]
correlation_matrix = df[['toeflScore', 'GRE_AWA', 'GRE_QUANT', 'GRE_VERBAL', 'GPA',
       'Publications', 'Research Experience', 'Work experience', 'Degree Encoded', 'Course Name Encoded']]

# Extract the correlation of 'university_name_encoded' with all other features

features = correlation_matrix
class_labels = df['University Name Encoded']


# Calculate Gini index for each feature
gini_values = []
for i in range(10):
    feature_values = features.iloc[:, i]
    gini, _ = calculate_gini_index(feature_values, class_labels)
    gini_values.append(gini)

# Normalize Gini index values
normalized_gini_values = normalize_gini_index(gini_values)

# Assign feature weights based on normalized Gini index values
feature_weights = normalized_gini_values

print("Feature Weights:", feature_weights)
```
